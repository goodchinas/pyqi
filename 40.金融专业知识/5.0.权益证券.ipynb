# 权益证券概述
## 种类
1. 普通股
>证券的最普通形式，代表了所有者权益。普通股东在公司破产或结业清算时，对于剩余资产的分配权排在债权人和优先股股东之后。
    
    1. 非累计制投票，每一股份仅有一票权利。
    2. 累计制投票，每一股份拥有与议案选项数相等的投票权。
2. 可赎回普通股，公司拥有以事先约定好的价格对股票进行赎回的权利。
3. 可回售普通股，股票持有人有权以事先约定的价格将股票卖回给公司。
4. 优先股，类似于债券，一般有固定的每期收益，一般没有选举权和被选举权。
    5. 累积优先股，通常承诺一个固定的股息，任何一个未支付的股息可以累积起来。
    6. 非累积优先股，只能按当年盈利分取股息。
    7. 参加分配，不参加分配的优先股
    7. 可转换优先股，允许优先股持有人在特定条件下，转化为一定数额的普通股。

## 不同权益证券的风险和收益
1. 权益证券投资收益包括，价格变动、股息收益、以及以外币计价的权益证券由于汇率变动带来的收益或损失。
2. 权益证券的风险通常用收益的标准差衡量。
3. 风险：优先股<普通股，累积优先股<非累积优先股，可回售<可赎回

## 市场价值和账面价值
1. 账面价值，资产负债表中公司的资产减去负债。
2. 市场价值，公司发行在外股票的总价，反映了投资者对公司表现的预期。
3. 股票的市价和账面价值很少一样。

## 净资产收益率（ROE）
$\large ROE=\frac{净利润}{账面价值的平均值}=\frac{净利润}{\frac{本期账面价值+前一期账面价值}{2}}\\\large ROE=\frac{净利润}{前一期账面价值}$

## 必要回报率和股本成本
股本成本是市场上股票预期的均衡总收益，实践中，通常用股利折现模型或者用资本资产定价模型来进行估计。在任何时点上，股票价格的下降将会增加股票的预期收益，相反，降低预期收益。

投资者估计的预期回报率高于可以接受的必要回报率，则可以投资。

公司的股本成本可以解释为在股票隐含风险下，投资者要求的最低的必要回报率。

# 行业分析和公司分析
## 行业分析的作用，行业分析和公司分析的关系
1. 行业分析为公司分析提供了一个框架和背景，通过行业分析可以更好地了解公司所面对的商业环境。
2. 主动投资管理中，通过行业分析决定哪些行业加仓，哪些行业减仓。
3. 绩效归因分析中，行业选择对应于择时，公司选择对应于选股。

## 公司分类方法
1. 通过公司所提供的产品和服务的类别将公司归类到合适的行业。
2. 根据公司业务对周期的敏感性进行分类，分为周期性行业和非周期性行业。
3. 应用统计学方法对行业进行分类，根据历史回报率相关性较高的公司分到一起，组与组之间的回报率相关性较低。

**商业分类标准**，指数提供商为公司进行分类，一般分为3个层次或者4个层次，最宽泛的是部门，紧接着是行业和子行业。

**政府分类标准**，政府提供行业分类方法和标准，通常是为了统计相关经济数据。

**周期性公司**的收入很大程度上依赖于商业周期所处阶段。例如，基础材料和加工、金融服务业。

**非周期性公司**提供的产品和服务的需求是相对稳定的。例如，医疗保健、公共事业。还可以进一步分为防御性行业和成长性行业。
1. 防御性行业最不被商业周期影响。
2. 成长性行业有很强的市场需求，也不收商业周期影响。

## 可比群体
可比群体，是指分析师用来进行价值比较的一组相似公司，具体来说，一个可比群体是由一些具有相似商业行为、需求驱动、成本结构驱动和资金使用方式的公司组成。构建可比群体的步骤：
1. 使用商业分类提供者提供的分类标准来确定哪些公司处于同一行业。
2. 查阅公司年报来观察他们是否认同他们的主要竞争者。
3. 查阅竞争者的年报来观察是否提供了其他竞争者。
4. 使用产业交易类的出版物来确定竞争者。
5. 确定可比的公司有相似的销售收入来源，相似的需求来源，并且处于相似的地理区域中。
6. 对任何非金融类公司的附属融资数据调整它的财务报表。

## 行业深度分析
行业深度分析需要涵盖以下内容：
1. 用行业、公司、竞争者、供应商和消费者等方面的数据，来评估宏观经济因素对行业趋势的影响；
2. 使用不同的方法和情景分析估计行业变量；
3. 与其他分析师对行业变量的进行比较，确保分析的准确性，同事能够发现潜在的行业价值被低估的情形；
4. 确定不同行业的相对价值；
5. 比较行业在一段时间内价值的变换，比确定他们在长期和不同的商业周期价值的波动性；
6. 业内战略集团与细分行业公司；
7. 定位行业所处生命周期阶段；
8. 定位行业在学习曲线中所处的位置；
9. 考虑影响行业的因素；
10. 考虑推动行业内竞争的因素；

## 行业战略分析的原则
五力分析框架
1. 行业内竞争者现在的竞争能力
2. 新进入者的威胁
3. 替代品的威胁
4. 购买者的议价能力
5. 供应商的议价能力

前两点对行业竞争环境的影响：
1. 较高的进入壁垒会减少竞争；
2. 更高的集中度会增加竞争；
3. 如果在一个行业中存在未被充分利用的能力，那么结果就会导致激烈的价格竞争；
4. 市场份额的稳定性降低了竞争的程度；
5. 客户购买的价格弹性越高，会导致更激烈的价格竞争；
6. 越成熟的行业，成长的速度就越慢；

## 公司分析中的因素
公司分析包括财务状况、产品或服务状况和竞争战略的分析。竞争战略是指公司如何应对外部环境的机会和挑战。策略可以是防守型，也可以是攻击型的（成本优势型策略和产品差别化策略）。

公司分析应该关注如下：
1. 公司整体的情况；
2. 行业的特点；
3. 产品的需求程度；
4. 生产成本；
5. 价格环境；
6. 财务比例；
7. 预测财务报表和公司估值；

# 股权估值
## 估值要点
1. 市价与预估的内在价值差距越大，投资者越会基于预估的内在价值进行投资，反之亦然。
2. 投资者对于使用的模型越有信心，约会投资于他认为被高估或低估的股票。
3. 投资者对于模型中使用的输入变量越有信心，越会投资于分析标的。
4. 市价应该作为内在价值的指标，必须思考为什么股票会被错误定价。
5. 投资被错误定价的股票，意味着投资者相信股价会向其内在价值靠拢。
6. 相对于受到较少关注的证券，被更多关注和研究的证券更容易被合理定价。

## 股权估值模型的类型

1. 现金流折现模型
    1. 股利折现模型（DDM），发放股利的现值。
    2. 自由现金流模型，可供股东支配现金的现值。
3. 乘数模型
    1. 股价和基本面数据的比值，如，市盈率
    2. 企业价值的乘数，如，企业价值和息税折旧摊销前利润或企业收入的比值

## 现金流折现模型
### 股利折现模型（DDM）
DDM模型的理念是股票的内在价值等于未来股利的现值。
公式：

$\large V_0=\frac{V_T}{{(1+k_e)}^T}+\sum_{t=1}^{T}\left [\frac{D_t}{{(1+k_e)}^t}\right]\\where:\\V_0:股票现价\\D_t:t时刻股利\\k_e:普通股的必要回报率$

DDM模型最常见的形式是持有期是无限期，在计算时，现值就等于所有与未来股利的折现值，股票的终值取零。

### 自由现金流模型
自由现金流模型，是将DDM模型中股利，用FCFE(自由现金流)替换得来。FCFE代表了股东潜在能收到的现金流，表示公司支付股利的能力，适用于不发放股利的公司。

$FCFE=\underbrace{净收入-折旧-周转资本的增加值}_{\large 经营性现金流}-固定资本投资\underbrace{-债务本金支付+新发行的债务}_{\large 净借款}$

**资本资产定价模型（CAPM）**

$E(R_i)=R_f+\beta_i\left[E(R_{m})-R_f\right]\\R_f:无风险利率\\E(R_m):市场的预期收益率\\\beta_i:贝塔系数\\E(R_i):贴现率（必要回报率）$
### 不可赎回、不可转换的优先股的内在价值
优先股的股利通常是固定且永久的。

$\large V_0=\frac{V_T}{{(1+k_e)}^T}+\sum_{t=1}^{T}\left [\frac{D_t}{{(1+k_e)}^t}\right]=\frac{D_p}{k_p}\\where:\\D_t=D_p,t\in[1,T]\\k_e=k_p$

### Gordon增长模型
假设股利的增长率是一个常数。

$\large V_0=D_0\sum_{t=1}^\infty{(\frac{1+g_c}{1+k_c})}^t=D_0(\frac{1+g_c}{k_e-g_c})=\frac{D_1}{k_e-g_c}\\where:\\g_c:股利增长率$

假设：
1. 股利是衡量股东财富的有效指标
2. 股利增长率$g_c$和必要回报率$k_e$永远不变
3. $k_e$必须要大于$g_c$。否则得到结果无意义

预测股利增长率：
1. 用股利增长率的历史数据；
2. 用行业股利增长率的中值；
3. 预测一个可持续的增长率；
>$可持续增长率=\underbrace{ 留存率}_{\large 1-股利发放比率}\times ROE$

### 两阶段股利增长模型
$\large V_0=\sum_{t=1}^{T}\left [\frac{D_t}{{(1+k_e)}^t}\right]+\frac{D_{T+1}}{{(k_e-g_c)(1+k_e)}^T}\\where:\\T+1及以后股利增长率固定为g_c.$


**适用性：**
1. gordon增长模型，适用于稳定的、成熟的、非周期的、发放股利的公司。
1. 两阶段股利增长模型，适用于现在经历高增长、之后回到稳定增速的公司，或者暂时低速增长、之后回归到稳定增速的公司。
1. 两阶段增长模型的变体可以有三个阶段，分为增长、过渡和成熟。
1. 当对未来股利预期不确定，基于FCFE的估值模型比较合适。

## 乘数模型
股利增长模型对于输入变量敏感。
### 价格乘数
价格乘数，由价格与基准价值相比而得，价格与基准价值可以基于一个指数、行业或者行业中与公司类似的一组公司。常用的价格乘数有：
1. $\large P/E=\frac{价格}{每股收益}$
2. $\large P/S=\frac{价格}{每股销售额}$
3. $\large P/B=\frac{价格}{每股账面价值}$
2. $\large P/CF=\frac{价格}{每股现金流}$

### 资产乘数
$企业价值（EV）=普通股和优先股的市价+债务的市价-现金和短期投资$

企业价值乘数，由企业价值（EV）与基准价值相比而得，EBITDA(利息、税、折旧前利润)是最常用的计算EV乘数的分母，也可以使用经营性收入。
### 估值
1. 统计行业或可比公司组的乘数
2. $V=行业乘数\times 当期或平均基础价值$

### 对比
1. 横向与行业或可比公司；
2. 纵向不同时间的对比；
