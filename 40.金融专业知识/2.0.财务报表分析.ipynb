# 财务报表分析
## 财务报告和财务报表分析的作用

公司通过准备和出示财务报告的方式将其财务表现展示给投资者、债权人和其他利益相关者。

财务报表的目标是提供关于一个实体的财务状况，经营情况以及现金流量状况的变化的信息，从而有助于广泛的使用者做出经济决策。

财务分析的目标就是运用公司财务报表中的信息，以及其他相关的信息，来做出经济决策。

## 财务报表的内容以及作用
### 资产负债表
1. 资产(Asset)：是经济实体掌控的资源；
2. 负债(Liabilities)：经济实体欠债权人的金额；
4. 股东权益(Equity)：是一个经济体中资产减去负债后剩余净资产的权益；

$资产=负债+所有者权益,A=L+E$

### 综合利润表
列示一个经济体一定时期的经营业绩，解释了在两个资产负债表日之间资产、负债、和所有者权益的变动。利润表包括：
1. 收入
2. 费用
3. 损益

### 股东权益变化表
是权益投资者在一段时间内对公司的投资的数量的变化；

### 现金流量表
反应经营、投资、筹资活动的现金流入流出情况。分类如下：
1. 经营活动现金流
2. 投资现金流
3. 筹资现金流

### 财务报表附注
1. 提供关于会计方法和所使用的会计假设和会计估计方法的信息；
2. 必须接受审计
3. 提供诸如商业收购或处置、法律事件、雇员福利计划、或有事项或或有承诺、重要客户、关联方交易和公司分支机构等方面的信息。

### 管理层讨论与分析
从一个企业的角度对财务业绩和公司状况的一个财务披露。

### 审计报告
审计是对一个经济实体的财务报表独立的意见。注册会计师经常执行审计的任务，检查财务报表及相应的支持数据。审计的目的是使审计人员能够对财务报表的公平性和可靠性提出意见。

标准的审计报告包括：
1. 审计人员必须保持独立客观的身份；
2. 审计人员确保报表符合要求，并且会计原则和会计估计方法的选择是合理的；
3. 无保留意见，有保留意见，否定意见；

# 财务报告机制
## 财务报表要素和科目的关系
指资产、负债、所有者权益、收入和费用的的划分。科目是指在每种要素中的各种交易的具体记录。在财务报表中，会计科目一般以组的形式出现。例如“存货”或是“应付账款”。会计科目表是组成五个会计要素的会计科目的详细列示，并以条目的形式显示在财务报表中。

会计科目按照会计要素分类：

**资产**是公司的经济资源，包括：
1. 现金及现金等价物
1. 应收账款
1. 存货
1. 金融资产
1. 预付款
1. 财产，厂房，设备
1. 以股权投资的方式投资的联营公司
1. 递延资产
1. 无形资产

**负债**是债权人对公司资源的要求权，包括：
1. 应付账款和应付货款
1. 金融负债
1. 预收收入
1. 应缴所得税
1. 长期负债
1. 递延负债

**所有者权益**是所有者对公司资源的除**负债**后剩余要求权，即资产超过负债的部分，包括：
1. 实收资本
1. 资本公积
1. 盈余公积
1. 其他综合收入

**收入**即经济资源的流入，包括：
1. 销售收入
1. 收益
1. 投资收入

**费用**即经济资源的流出，包括：
1. 已出售货物的成本
1. 销售费用
1. 折旧和摊销
1. 税收支出
1. 利息费用
1. 损失来自于公司非主营业务的资产或是所有者权益的减少

## 会计等式

$资产=负债+\underbrace{所有者权益}_{\large 所有者投入的资本+\underbrace{期末留存收益}_{期初留存收益+收入-费用-股利支出}}$

## 复式记账法
是会计等式恒等要求复式记账法，即，每一笔交易要**至少在两个会计科目中记录**。例如，资产的增加，必然会以另一项资产的减少，或另一项负债或所有者权益增加为平衡。

## 权责发生制

收入和费用并不总是与现金的收到和支付的清偿同时记录的。权责发生制的原则要求，当公司赚得收入的时候就记录收入，当公司发生费用的时候就记录费用，而不管是否真正被支付。权责发生制可分为以下四类：
1. 预收收入。公司在向客户提供商品或服务之前收到现金，现金增加，预收收入（负债类科目）也同时等额增加了。当公司提供了商品或服务的时候，收入增加，负债减少。
1. 应计收入。公司在收到现金支付之前提供商品或劳务，收入增加，应收账款（资产类科目）也同时等额增加了。当客户支付了现金，现金增加，应收账款减少。
1. 预付费用。公司提前支付确定金额的费用。现金（资产）减少，预付费用（资产）增加。当费用实际发生的时候，预付费用减少，费用增加。
1. 应计费用。公司拖欠已经发生的费用。费用增加，应计费用（资产）也增加。当公司支付了现金之后，现金减少，应计费用也减少。

权责发生制要求当最早的事件一发生就进行会计记录，并要求当交易完成时有一个或多个冲销会计记录。对于预收收入或预付费用，现金交易先发生，之后再对收入和费用做记录。对于应计收入和应计费用，收入和费用先记录，之后才发生现金的交易。

# 利润表分析
利润表反映的是一个公司在一定时期内的收入和支出情况。利润表有时也被称为“经营情况表”或“损益表”。利润表的等式为：$净利润=收入-费用$。
## 利润表构成
收入是销售商品和提供劳务的所得。收入减去估计的销售退回和折扣就是净收入。

支出是在产生收入的过程中出现的，包括销货成本，经营费用，利息和税收。支出通过其性质或功能分类。例如，把生产和管理中心的折旧费归为一类是按支出的性质来分类的，把关于生产的一切成本（如原料、折旧、人工等）都划分为销货成本就是按功能分类的。

利润表也包括了公司主营业务之外的盈余和损失。盈余和损失是由于经济利益的增加和减少所产生的。

$净利润=\underbrace{收入}_{主营收入+营业外收入+盈余}-\underbrace{费用}_{一般费用+其他费用+损失}$。

$毛利润=收入-一般费用$

$营业利润=毛利润-\underbrace{其他费用}_{销售费用+管理费用}$

## 收入确认
在权责发生制下，当取得的时候就确认收入，当应计的时候就确认支出。值得注意的是权责发生制不一定要和现金的收入和支出相一致。这样，公司可以通过提前或延迟确认收入，或者是加速或推迟确认支出来操纵利润。

### **收入**在以下货物交易环节中实现：
1. 风险和所有权的转移
1. 已经售出的货物已经不被控制和管理
1. 收入可以可靠计量
1. 将可能产生经济利益的流入
1. 成本可以可靠计量

### **收入**在以下服务交易环节中实现：
1. 收入可以可靠计量
1. 将会产生经济利益流入
1. 完成的进度可以计量
1. 已经发生和将要发生的成本可以准确计量

### **收入**还可以按如下标准确认：
1. 买卖双方之前确有交易发生
1. 货物已发生或者劳务已经提供
1. 价格已经确定或能够被确定
1. 卖方确信能够收回货款

### 具体的收入确认方法：

收入一般在发货时按照前面的确认标准来确认。然而，有些案例，收入有时候会在交货之前或之后确认。

#### **长期合同**（一般是建筑项目合同）：
1. 完工百分比法。当项目的成本和收入能够可靠计量时。收入、费用和利润根据合同完工的进度被确认。完工百分比是由目前的成本和预期总成本之比计算的。
2. 完工合同法。当项目的最终结果难以可靠计量或者是项目是短期的。收入、费用和利润当合同完成时才确认。

#### **分期收款销货**
当公司以融资的方式销售并在一个延长时期内收取货款时就是分期收款销货。
1. 如果可回收性确定，则收入在销售时就用普通的确认方法确认。对于利润，同样以普通的方法确认。
2. 如果可收回性不确定，那么就使用分期收款法。对于利润，当现金收到时就可以确认，利润等于收到的现金乘以总的预期利润占销售收入的百分比。
3. 如果可回收性特别不确定，就使用成本回收法。对于利润，只有在收到的现金超过成本时才能确认。

分期收款法和成本回收法一般在房地产出售时使用。
#### 易货交易
在易货交易中，交易双方交换商品或劳务而不是现金的支付，即向一方出售货物的同时向其买入相等价格的货物。这类交易隐含的问题是收入是否应该被确认。

原则：
1. 来自易货交易的收入只有在公司以前从这些商品和劳务中取得过现金支付的情况下才能被确认，并且可以用以前的交易经验来决定公允价值。
2. 来自易货交易的收入必须是基于其他方的相似的非易货交易的公允价值确认。

## 费用确认
费用是在会计期间以资产的流出或耗损或负债的发生为表现形式的经济利益减少，从而会造成所有者权益的减少。

在权责发生制下，费用的确认遵循配比原则，因此，产生收入的费用，在收入确认的同时被确认。

不是所有的费用都能直接与收入相匹配。这类费用被称为期间费用，例如，管理费用。

### 存货发出的确认
1. 实际计价法
2. 先进先出法。先行购入存货假设先行售出，先购入的存货的成本对应于本期先售出存货成本。近期购入存货的成本对应期末存货的价值。
3. 后进先出法。最近购入存货的成本对应本期售出存货的成本。期初存货的成本和本期早些时间购入存货的成本转到期末存货的价值上。
4. 平均成本法。每一单位存货成本的计算方法是可供出售的存货的总成本除以总的可供出售的存货的数量。

|方法|假设|已售存货|期末存货|
|:-|-:|-:|-:|
|先进先出|先购入存货先出售|先购入存货|后购入存货|
|后进先出|后购入存货先出售|后购入存货|先购入存货|
|平均成本法|各时期存货混合出售|所有存货平均成本|所有存货平均成本|

### 折旧费用的确认
长期资产的成本也必须与收入相匹配。长期资产会在超过一个会计期间的过程中提供经济收益。把成本分配到一项资产的使用寿命过程中叫做折旧（有形资产），减耗（自然资源）或分期摊销（无形资产）。
1. 直线折旧法。大多数公司在财务报告中用直线折旧法。直线折旧法每期确认的折旧费用相同。
2. 加速折旧法。然而很多资产在经济寿命的前期会产生更多的利益，在后期产生相对较少的利益。这种情况下，加速折旧对收入与费用的配比更合适。
>在资产使用寿命前期，加速折旧法相对于直线折旧法会产生较低的折旧费用。较低的折旧费用会提高净利润。在资产使用寿命的后期，正好相反，加速折旧法相对于直线折旧法会产生较高的折旧费用和较低的净利润。
3. 折旧余额法。把一个固定的折旧率应用到递减的账面价值中。最常见的折旧余额法就是双倍余额递减，这种方法用直线折旧法折旧率的两倍作为余额的折旧率。

### 摊销费用的确认
无形资产的摊销费用是将无形资产的成本分摊到它的使用期。
1. 大部分情况下采用直线摊销法，与直线折旧计算法相同；
2. 对于没有确定使用年限的无形资产是不能摊销的（例如商誉）。它们至少每年要做减值测试。如果这项资产的价值减少了，那么对应减少的等额费用，在利润表上确认。

### 坏账费用和质保费用的确认
如果公司赊销产品或为客户提供质保，配比原则要求企业估计坏账损失和质保费用。这样公司根据一定的销售期限确认此费用。

## EPS-每股收益
每股收益是公开上市的公司公布其利润情况的最常用的指标之一。

简单资本结构是不包含潜在的可以稀释每股收益的债券。具有简单资本结构的公司的资本仅包含普通股、不可转换债券和优先股。

复杂资本结构是包含潜在的可以稀释每股收益的债券。如期权、认股权证或可转换债券。

所有拥有复杂资本结构的企业必须同时报告其基本EPS和稀释EPS。
### 基本EPS
$\large基本每股收益=\frac{净利润-优先股股利}{平均发行在外的普通股}$

加权平均股数和EPS

$\large EPS=\frac{\sum^{n}{净利润}-\sum^{n}{优先股股利}}{\sum^{n}{普通股}}$

### 稀释EPS
* 可稀释每股收益的证券包括股票期权、认股权证、可转换债券或可转换优先股，这些证券被执行或转换成普通股后将会减小每股收益。
* 反稀释每股收益的证券是指执行或转换成普通股后能够增加每股收益的证券。
* 
$\large稀释每股收益=\frac{[净利润-优先股股利]+[可转换优先股股利]+[可转换债券税后利息]}{[加权平均普通股数]+[转换可转换优先股后增加的普通股数]+[转换可转换债券后增加的普通股数]+[执行股票期权后新发行的普通股数]}$

### 比例
#### 比例格式利润表
比例格式利润表由百分比的形式呈现利润表的各项指标，在对企业进行纵向（不同时期）和横向（同类企业）对比时，它消除了企业规模大小的影响
#### 利润率
利润率可以很快得出企业的盈利能力。

$毛利率=\frac{毛利润}{收入}$

$净利率=\frac{毛利润-其他费用-损失}{收入}$

# 资产负债表分析
资产负债表，也叫公司财务状况表，是公司财务状况在一个时点的表现（利润表是一个时期的表现），由资产、负债和所有者权益组成。

资产负债表可以用来测试公司的流动性、偿付能力、对股东的盈利贡献能力。从公司的角度来说，流动性是对短期债务的应付能力，偿付能力显示了他对长期债务的应对能力。

由于多种计量方法的存在，资产负债表中的要素不能被理解为公司的市场价值或固有价值。

一般情况下，公司分别列出流动和非流动资产，流动和非流动负债。流动对比非流动的形式成为分类资产负债表，这对测试公司的流动性是有用的。

## 流动资产
1. 现金及现金等价物
1. 可供出售证券
1. 应收账款
1. 存货
1. 其他流动资产

## 非流动资产
除流动资产，不能在一年或一个经营周期中转化为现金或用完。非流动资产体现了公司投资活动的信息。
1. 地产、厂房、设备
1. 投资资产
1. 无形资产
1. 金融资产

## 流动负债
流动资产减去流动负债就是营运资产。营运资产不足反映了流动性问题。过多的营运资产则反映了资产利用的效率不高。
1. 应付账款
1. 应付票据
1. 应计费用
1. 预收收入

## 非流动负债
除流动负债，非流动负债提供了公司长期融资活动的信息。
1. 长期金融债
1. 递延所得税负债

## 股东权益
1. 实收资本
1. 优先股东权益
1. 留存收益
1. 库藏股票
1. 累积的其他总综合收入

### 股东权益变动表

## 比率
### 流动性比率：
1. $流动比率=\frac{流动资产}{流动负债}$
2. $速动比率=\frac{现金+短期证券+应收账款}{流动负债}$
3. $现金比率=\frac{现金+短期证券}{流动负债}$

### 清偿能力比率
1. $负债所有者权益比率=\frac{负债}{所有者权益}$
1. $负债率=\frac{负债}{资产}$
1. $财务杠杆=\frac{资产}{所有者权益}$

### 比率分析的局限性
1. 不同的会计估计和财务标准
2. 公司行业的差别
3. 资产负债只是时间点的数据

# 现金流量表分析
## 现金流量表
现金流量表是财务报表的重要组成部分。利润表是基于权责发生制，净利润可能并不能代表经营所产生的现金。现金流量表是按收付实现制基础编制的，能够提供利润及其他财务数据不能提供的信息：
1. 提供在一个会计年度企业的现金收入和支出的相关信息。
2. 提供企业经营、投资、融资活动方面的相关信息；

如果一个企业想做出下一步的投资决策，那么其必须从现金流量表上获得相关信息：
1. 企业的日常经营项目是否产生足够的现金以支持企业的持续经营。
2. 是否产生足够的现金已偿还到期债务。
3. 企业是否需要额外的融资。
4. 无法预期的一些或有债务是否能支付。
5. 当有新的投资机会出现时，企业是否有足够的现金能把握住这次机会。

## 条目分类
现金流量表的条目来自于两个方面
1. 利润表上的条目变化；
2. 资产负债表上的条目变化；

按现金的收取和支付划分：
1. 经营活动现金流（CFO-cash flow from operations），从有关流动资产和流动负债的变动中产生，包括从影响公司净利润的交易中产生的现金流入和流出。
2. 投资活动现金流（CFI-cash flow from investing），从有关非流动资产的变动中产生，包括在收购或处置长期资产和某些投资过程中产生的现金流入和流出。
3. 融资活动现金流（CFF-cash flow from financing），从有关非流动负债和所有者权益的变动中产生，包括从影响公司资本结构的交易中产生的现金流入和流出。

**注：**
1. 取得的负债和权益投资（而不是交易证券）以及向其他方的贷款作为**投资活动记录**。
3. 支付给股东的股利是**融资活动**。
4. 非现金投资和融资活动不记录在现金流量表中，非现金交易必须在财务报表附注或现金流量表附表中披露。

## 列示现金流量表的方法
**直接法：**

直接法，以权责发生制为基础，利润表中的每个条目都转化为现金的收入与支出。权责发生制，收入与费用确认的时间可能会与现金流入流出的时间不同。收付实现制，收入与费用只有在现金得到支付时才确认。直接法把利润表由权责发生制转化为收付实现制。

**间接法：**

间接法，通过对那些影响净利润但又非现金的交易的调整，把净利润转化为经营现金流。调整包括**减去**非现金支出（折旧和摊销），非经营条目（营业外损益）以及来自权责发生制的资产负债表的变化。

间接法计算的起点是**净利润**，利润表的最下端，而直接法计算的起点是利润表的最上端。两种方法下得到的经营现金流基本上是一样的，只是表现形式不同。

## 现金流类比率
自由现金流（FCF-free cash flow）度量的是公司可以自由支配的现金流，经常用于估值的目的。
>$自由现金流=经营活动现金流(CFO)-保持当前公司生产能力必须支出的现金流$

两种常用的度量方法是公司自由现金流（FCFF）和权益自由现金流（FCFE）。
>$FCFF=\underbrace{净利润+非现金支出-营运资本投资}_{经营现金流(CFO)}+[利息费用*(1-税率)]-固定资本投资$
$FCFE=经营现金流(CFO)-固定资本投资+发行的债权-偿还的负债$

经营业绩比率：
>$现金收入比率=\frac{CFO}{净收入}\\现金资产比率=\frac{CFO}{平均总资产}\\现金权益比率=\frac{CFO}{平均总权益}\\现金收益比率=\frac{CFO}{经营收益}\\每股现金流=\frac{CFO}{总股数}$

偿还能力系数：
>$负债保障倍数=\frac{CFO}{总负债}\\利息保障倍数=\frac{CFO+支付的利息+支付的税收}{支付的利息}\\再投资倍数=\frac{CFO}{用于购买长期资产的现金}\\负债偿还倍数=\frac{CFO}{长期负债现金支出}\\股利支付倍数=\frac{CFO}{股利支付}\\投融资比率=\frac{CFO}{投融资活动的现金流出}$

# 财务分析技术
把财务报表数据转换为用于分析的方法，包括比率分析，通用格式分析，图表分析和回归分析。
## 比率分析
比率分析可以用来表述数据之间的内在关系，以及进行行业内对比。比率分析对于确定问题有用，但不是回答问题。可以用于以下情况：
1. 估计项目的未来收益和现金流
2. 估计公司的能力（成长、偿付、应对突发）
3. 评价管理水平
4. 估计公司的变化和行业的发展
5. 比较同行的竞争力

比例分析的局限性：
1. 财务比率单独使用并没有意义，它们只有与本公司历史业绩比较或与其它公司进行比较的时候才有用
2. 当财务比率在不同公司之间进行比较时，会出现会计计量方法不同而造成比率的比较不精确的困难
3. 考察跨行业的大型公司的业绩时，要找到一个比较合适的行业标准来作为比较依据会显得有些困难
4. 仅从单一系列指标进行分析不能做出结论，结论可能是片面的，比率只有与其他比率进行比较时才能做出整体而全面的结论
5. 如果为一个比率设定一个标准比率作为比较的衡量标准是困难的，大多数情况下，设定一个可以接受的区间

## 通用格式分析
通用规则财务报表把资产负债表和利润表一般化，以比较不同公司间或一个公司不同时期的财务表现：
1. 垂直通用规格资产负债表把所有资产负债表中的科目表示成总资产的百分比
2. 垂直通用格式利润表把所有利润表中的科目表示成销售收入的百分比

## 分类
财务比率根据提供的公司信息不同分类：
1. 业务活动比率
2. 流动性比例
3. 清偿能力比率
4. 盈利能力比率
5. 价值比率

## 业务活动比率
>$应收账款周转率=\frac{年净收入}{年平均应收账款}\\应收账款平均收款期=\frac{365}{应收账款周转率}\\存货周转率=\frac{销售成本}{平均存货}\\存货平均处理周期=\frac{365}{存货周转率}\\应付账款周转率=\frac{销售成本}{平均应付账款}\\应付账款平均付款期=\frac{365}{应付账款周转率}\\总资产周转率=\frac{销售收入净值}{平均总资产净值}\\固定资产周转率=\frac{销售收入净值}{平均固定资产净值}\\营运资本回收率=\frac{收入}{平均营运资本}$
## 流动性比率
>$流动比率=\frac{流动资产}{流动负债}\\速动比率=\frac{现金+短期证券+应收款}{流动负债}\\现金比率=\frac{现金+短期证券}{流动负债}\\防御区间比率=\frac{现金+有价证券+应收账款}{平均每日支出}\\现金转换周期=应收账款平均收款期+存货平均处理周期-应付账款平均付款期$
## 清偿能力比率
>$负债权益率=\frac{总负债}{所有者权益}\\资产负债率=\frac{总负债}{总资产}\\财务杠杆比率=\frac{平均总资产}{平均总权益}\\利息保障倍数=\frac{利息税前利润}{利息费用}\\固定费用偿付比率=\frac{利息税前利润+支付的租金}{支付的利息+支付的租金}$
## 盈利能力比率
>$净利润率=\frac{净利润}{销售收入}\\\Large 税后净利润(EAT)=\underbrace{税前利润(EBT)}_{\underbrace{营业利润(EBIT)}_{\underbrace{毛利润}_{销售收入净额-销货成本}-营业费用}-利息}-所得税\\毛利润率=\frac{毛利润}{销售收入净额}\\营业利润率=\frac{营业利润|息税前利润}{销售收入}\\资产收益率(ROA)=\frac{净利润}{平均总资产}\\普通股收益率=\frac{净利润-优先股股息}{平均普通股}$

## 杜邦分析-资本收益率(权益净利率(ROE))的分解
杜邦财务分析体系，是一种分解权益净利润(ROE)的方法。其运用最基本的数学方法将ROE分解为不同比率的乘积形式，分为两种体系：
>$\large权益净利率(ROE)\\\large=\frac{净利润}{所有者权益}=\frac{净利润}{销售收入}\times\frac{销售收入}{所有者权益}=销售净利率\times 资本周转率\\\large=\frac{净利润}{销售收入}\times\frac{销售收入}{总资产}\times\frac{总资产}{所有者权益}=销售净利率\times总资产周转率\times权益乘数\to\color{\red}{(三部分体系)}\\\large=\frac{净利润}{税前利润(EBT)}\times\frac{税前利润(EBT)}{营业利润(EBIT)}\times\frac{营业利润(EBIT)}{销售收入(营业总收入)}\times\frac{销售收入}{总资产}\times\frac{总资产}{所有者权益}\\\large=税收负担\times利息负担\times营业利润率\times总资产周转率\times权益乘数\to\color{\red}{(五部分体系)}$

杜邦分析的作用不在于计算ROE，而在于分析ROE，考察什么原因使ROE变化。

## 价值比率
>$市盈率=\frac{股价}{每股收益}\\市净率=\frac{股价}{每股净资产}\\市现率=\frac{股价}{每股自由现金流}\\市销率=\frac{股价}{每股销售收入}$

## 经营风险
>$销售收入波动系数=\frac{销售收入标准差}{平均销售收入}\\经营收入波动系数=\frac{经营收入标准差}{平均经营收入}\\净利润波动系数=\frac{净利润标准差}{平均净利润}$
