# 资产类别

## 资产类别的定义
一个资产类别是指一组具有相似特征的资产。

### 特征
1. 风险
2. 流动性
3. 期望收益

### 资产
1. 金融资产
2. 非金融资产

## 主要的资产类别
1. 主要
    1. 现金及现金等价物
    2. 权益资产（股票）
    3. 固定权益证券
2. 其他
    1. 衍生金融产品
    2. 外汇
    3. 不动产
    4. 基础设施
    5. 商品

### 现金及现金等价物

#### 现金
包括企业库存现金以及可以**随时用于支付**的存款。

#### 现金等价物
1. 期限短
2. 流动性强
3. 易于转换为可确认金额的现金
4. 价值变动风险很小

#### 可作为现金等价物的定期存款
1. 非受限
2. 管理层持有意图是为了满足短期内对外支付的流动性需求，而不是为了获取利息收入或投资收益为主要目的

### 权益资产
公司权益通常称为“股东权益”，是指假设公司清算后，全部资产减去全部负债后，可供股东分配的余额（股东权益=总资产-总负债）;

作为资产类别的一部分，权益资产除了包括直接的股东权益之外，还包括价值主要依赖于股东权益资产价值变动的资产，如存托凭证、股票型基金等。

权益资产包括上市权益资产和非上市权益资产：
1. 上市权益资产：指在交易所公开交易的股票，以及主要价值依赖于股票价值变动的资产；
2. 未上市权益资产：指未在交易所公开交易的企业股权，以及主要价值依赖于未上市企业股权价值变动的资产。

### 固定收益证券
固定收益证券持有人可预期获得定期的利息收益，并可获得本金偿付
1. 债券属性
2. 事先确定利息以及本金支付的金额及时间
3. 主要包括国债、金融债、公司债、企业债等，未归入现金等价物的定期存款，可归入固定收益证券

## 区分资产类别的价值

### 分散化与降低非系统风险
1. 不同类别的资产，通常呈现弱相关或负相关
2. 分散于不同资产类别之间，比分散于同一类别不同标的资产之间，降低非系统风险效果更好

### 风险偏好于风险承担
1. 不同的投资者有着不同的风险偏好
2. 不同资产类别通常有着不同的风险水平
    1. 货币市场基金
    2. 债券基金
    3. 股票基金
    4. 私募股权基金
3. 不同资产类别通常也有着不同期望收益水平

### 投资策略
不同的投资策略可能需要投资于不同的资产类别
> eg:beta策略通常仅投资于某一个资产类别（如股票）；儿alpha策略则更多涉及到同时投资于不同类别的资产；

### 资产配置
1. 确定风险偏好和风险胃口；
2. 判断不同资产类别的风险和期望收益；
3. 确定不同资产类别的配置比例；（有研究表明第三部分影响要大于第四部分）
4. 选择具体的标的资产，择时投资；

# 货币的时间价值
投资学三大投资工具：
1. 货币的时间价值
2. 资产的定价
3. 风险的分析和管理

## 货币的时间价值？
1. 金融分析中，同样金额的资金，发生在不同时间，其实际价值是不相同的。
2. 货币的时间价值，是指货币的价值随时间推移而发生的变化。

## 货币时间价值的来源？
1. 不耐：人们有“现时收入优于未来收入”的偏好；
2. 机会：早获得的资金，可以使用投资机会获得投资收益；
3. 风险（未来的不确定性）：
    1. 信用风险；
    2. 通货膨胀风险

## 货币的时间价值的核心要素
1. 利率与贴现率
    1. 利率：资金增值与投资资金的价值之比；
    2. 单利和复利；
    3. 风险利率（一般国债利率）与无风险利率；
    4. 名义利率（资金名义增幅）=实际利率（资金购买力实际情况，剔除通货膨胀因素）-通货膨胀率；
    5. 即期利率与远期利率（未来利息再投资）：
    6. 利率与贴现率（未来资金的贴现计算出来的利率）的关系：：
        1. 利率通常以借贷关系为背景；
        2. 利率常被作为贴现率；
        3. 但贴现率不都是利率；
2. 终值（FV）：一笔或多笔现金流期末的价值；
3. 现值（PV）:
4. 年金（Annuity）：

## 货币时间价值的计算
1. 单利终值：$FV=PV\times(1+i \times t)$
2. 单利现值: $PV=\frac{FV}{1+ i \times t}$
3. 复利终值：$FV=PV\times {(1+i)}^t$
4. 复利现值：$PV=\frac{FV}{{(1+i)}^t}$

## 货币时间价值的计算-举例
小张新买一套房，放款总额300万元，首付比例为40%，剩余未贷款，贷款期限为30年，利率为4.9%，按月等额本息还款：
1. 贷款金额为$300\times (1-.4)=180$万元
2. 贷款期限为$30\times 12=360$期
3. 每期利率为$\frac{4.9\%}{12}=.408%$
4. $180=\sum_{t=1}^360{\frac{M}{{(1+i)}^t}}$，解的M=9553.08元
5. 总还款为$9883.08\times 360=3439108.8$元
6. 因时间推移而增加的价值约为163.9万元
7. 实际情况是住房抵押贷款每年调整贷款利率，而不是一直不变的

# 财务报表分析基础

## 企业基本活动
1. 业务经营:采购、生产、销售商品，或提供服务
2. 筹资： 债务融资、股权融资
3. 投资: 短期长期投资

## 财务报表基础
1. 主要包括资产负债表、利润表、现金流量表
    1. 资产负债表报告截至某一时间点企业的资产、负债和所有者权益状况；
    2. 利润表报告某一段时期企业的收入、成本、盈亏情况；
    3. 现金流量表报告谋一时期内企业由于经营、筹资和投资活动导致的现金变动情况；
2. 会计的基础工作，是按照符合会计准则的会计政策，对企业生产经营及投融资活动进行簿记，并汇总为财务报表；
3. 审计的基础工作，是核实企业簿记及汇总是否符合会计准则要求，是否客观真实；
4. 财务报表应发挥的作用，是客观、准确、充分的反映企业生产经营状况和经营财务结果；


资产负债表目录项目结构（T型结构）：
1. 资产:流动、非流动
2. 负债和所有者权益
3. 所有者权益=资产-负债，恒等式

利润表结构
1. 营业收入、营业成本、营业利润

现金流量表结构
1. 经营活动产生、投资活动产生、筹资活动产生的现金流量
2. 汇率变动对现金及现金等价物的影响、期初、期末


## 财务报表分析基础
1. 财务报表分析
> 指通过对企业财务报表进行解析，挖掘企业生产和经营的信息，为评估企业的生产经营业绩和财务状况提供帮助
2. 财务报表分析的功能
    1. 价值评估：财务报表分析可服务于对企业进行估值；
    2. 风险发现：通过财务报表分析，可以识别企业可能存在的风险；
3. 财务报表分析的主体和主要目的：
    1. 股东：关注企业价值创造和股东权益及收益
    2. 债权人：关注企业偿债能力
    3. 经理人：关注企业经营状况和经营业绩
    4. 市场投资者：关注企业投资价值

### 财务报表分析的内容和方法：偿债能力分析
1. 总体/长期偿债能力：
    1. 资产负债率=$\frac{总负债}{总资产}$
    2. 杠杆比率（权益乘数）=$\frac{总资产}{所有者权益}$
    3. 负债权益比=$\frac{总负债}{所有者权益}$
    4. 利息保障倍数=$\frac{EBIT：息税前利润}{利息支出}$
2. 短期偿债能力
    1. 流动比率=$\frac{流动资产}{流动负债}\geq 2$
    2. 速动比例=$\frac{流动资产-存货}{流动负债}\geq 1$

### 财务报表分析的内容和方法：经营效率分析
1. 存货周转率
2. 存货周转天数
3. 应收账款周转率
4. 应收账款周转天数
5. 总资产周转率

### 财务报表分析的内容和方法：成长性分析
1. 总资产增长率
2. 净资产增长率
3. 销售收入增长率
4. 利润增长率
5. 市值增长率


### 财务报表分析的内容和方法：成长性分析
1. 销售利润率
2. 资产收益率
3. 净资产收益率

### 杜邦分析法
权益净利率=销售净利率*总资产周转率*权益乘数

## 财务报表分析服务于企业估值
1. 成本法-清算价值法
2. 收益法-贴现现金流量法
3. 市场法
    1. 市盈率法
    2. 市销率法
    3. 市净率法

## 财务报表分析服务于风险识别
1. 财务报表是一个完整封闭的系统，勾稽逻辑严谨
2. 可利用逻辑分析识别财务造假嫌疑
    1. 账面现金很多，但利息收入很低
    2. 账面现金很多，但有息负债很高
    3. 当期利润很高，单经营活动产生的现金流流入很低
    4. 账面现金减少，同时其他应收款增加
    5. 其他应收款增加，增量与当期利润可比
    6. 在建工程增加，增量和当期利润可比
    7. 毛利率超过行业平均，当研发费用占比/人均工资低于行业平均水平
 

# 股票价值分析


## 基本逻辑
1. 股票的收益来源
    1. 当期收益&预期收益
    2. 分红收益&股票增值收益
2. 上市公司股权价值的影响因素
    1. 上市公司的经营：当期业绩、未来业绩预期
    2. 市场扰动因素：舆情、投资者行为等
3. 分析方法：基本面、技术分析

## 基本面

### 宏观分析
1. 宏观经济指标
    1. GDP：需求端（投资、出口、消费、政府支出）、供给端（农业、工业、服务业等）
    2. 工业总产值
    3. 社会零售品销售总额
    4. 通货膨胀
    5. 利率、汇率、失业率
    6. 消费信心指数、投资信心指数、PMI
2. 同步指标（GDP、工业总产值，社会零售品销售总额）
3. 先行指标（X信心指数、工作小时数、发电量、用电量等）
4. 滞后指标（失业率）
5. 经济周期（繁荣、衰退、萧条、复苏）
6. 宏观经济政策
    1. 财政政策
        1. 政府支出
            1. 基础设施建设：铁公鸡
            2. 公共事业：科教文卫保
            3. 行政管理
        2. 政府收入
            1. 税
            2. 费
            3. 债
    2. 货币政策
        1. 公开市场操作
        2. 利率
        3. 存款准备金

### 行业分析
1. 行业生命周期
    1. 初创期
    2. 成长期
    3. 成熟期
    4. 衰退期
2. 行业景气度
    1. 行业景气指数
    3. 经理人信心指数

### 公司分析
1. 价值与价格
    1. 如果价值高于价格，则买入；
    2. 如果价值低于价格，则卖出
2. 公司价值判断：股票估值
    1. 内在价值法:关注公司自身
        1. 股利贴现模型（未来所有股利贴现现值综合）
        2. 现金流贴现模型
        3. 经济附加值模型
    2. 相对价值法：可比，相同或者相近的行业公司
        1. 市盈率
        2. 市净率
        3. 市销率
        4. 市现率

## 技术分析
1. 市场行为涵盖所有信息
2. 股票价格沿趋势运动
3. 历史会重演

### 基础工具
1. 价：高开低收
2. 量
3. K线图
4. 均值

### 常用方法
1. 道氏理论
    1. 趋势分析：股票随市场趋势同向变化
    2. 长期、中期、短期趋势
2. 过滤法则：当股票价格变化突破某一百分比，就进行买或卖的交易
3. 相对强度：交易近期走势明显强于大盘的股票
4. 价量分析
    1. 放量上涨，买入
    2. 放量下跌，卖出
5. 有效市场理论与技术分析有效性（fema有效性，国内处在弱势到半强式的状态）
    1. 弱势有效（）
    2. 半强式有效（）
    3. 强势有效

# 债券类型


## 债券市场概述
1. 债券通常又称为固定收益证券
2. 债券是标准化的债务融资工具，债券发行人和持有人指甲泥是债务人和债权人的关系
3. 债券能够为持有人提供固定数额或根据固定公式计算出的现金流

## 债券种类

### 发行主体
1. 政府债券：国债、地方债、央行票据
2. 政策性银行债券/金融债：
    1. 政策性金融债（国开发、进出口，农开行）
    2. 商业银行债券
    3. 特种金融债
    4. 非银行金融机构债券
    5. 证券公司债
    6. 证券公司短期融资
3. 信用债券
    1. 企业、公司、中期票据、短期融资券、大额存单

### 发行主体-实践情况：

#### 利率债
包括：国债、地方债、央行票据、**政策性金融债**
1. 参与者多为专业机构
2. 银行间市场是债券市场主体
3. 市场走势的决定因素多为宏观基本面，基本不受人为干扰
4. 亏损和收益相对可控
    1. 市场利率的变化具有周期性，随着时间的推移，当前的未实现收益最终会逐渐消失，甚至会转为亏损
    2. 市场利率的变化是有限度的，因此债券价格的上涨和下跌也会被控制在一定范围内

#### 信用债
除利率债其他发行主体的债
1. 信用债的大趋势与利率债趋势正相关
    1. 利率的分析也是判断信用债走势的基础
2. 信用债和利率债的最重要差别体现在信用利差
    1. 信用债发行主体资质不同，信用利差不同
    2. 同一发行主体，其经营状况发生了变化，信用资质也发生变化
    3. 对信用债的投资，不能忽略对发行主体的个别分析
3. 同为公司或行业研究，信用债投资和股票投资的着眼点存在差异
    1. 信用债投资更关注企业的偿债能力，更希望看到企业能够平稳经营；
    2. 股票投资更关注企业的发展潜力，更希望看到企业快速增长，更喜欢“故事”
4. 信用债投资，需要机构具有强大的信用评估能力
    1. 个人和小机构投资信用债，多依赖于评级机构
    2. 公开评级大多是企业为发债聘请机构所做，其结果仅能作为参考
5. 信用债的优势
    1. 票息高：高票息带来高收益，同时也带来较好的全价保护
    2. 债券熊市一般出现于经济繁荣时期，企业经营状况好，信用资质提升，信用利差下降；因此信用债在债券趋势性下跌前期跌幅一般小于利率产品
6. 信用债的劣势
    1. 流动性差，需要较强的交易能力
    2. 资信状况需要准确评估
    

### 偿还周期
1. 短期（原1年以内，实际1-3年）
2. 中期（原1-10年，实际3-5年）
3. 长期（原1-10年，实际5年以上，10年以上为超长债）

### 计息方式
1. 固定
2. 零息
    1. 折价发行，面额偿付
    2. 多为1年以内短期利率产品
3. 附息债券
    1. 面额发行，面额偿付
    2. 多数债券的发行方式
        1. 固定利率
        2. 浮动利率

### 付息方式
1. 息票债券：附息债券
>债券发行时规定，在债券存续期内，在约定的时间以约定的利率按期向债券持有人支付利息的中、长期债券
2. 贴现债券：零息债券
>无息票债券或零息票债券，在发行时不规定利率，不附息票，折价发行，到期按债券面额支付

### 嵌入条款
1. 可赎回
    1. 发行人的权利
    2. 利息更高
1. 可回售
    1. 受益人是债券持有人
    2. 利息更低
1. 可转换/可交换
    1. 可转换：可转换成股票
    2. 可交换：上市公司股东通过抵押其持有的股票给托管机构进而发行的公司债券

## 短期回购协议
1. 回购协议功能
    1. 央行公开市场操作
    2. 商业银行流动性和资产结构管理
    3. 非银行金融机构通过证券回购协议实现套期保值、头寸管理、资产管理、增值等
2. 回购协议市场
    1. 以债券回购协议为主
    2. 场内交易：上交所和深交所开办的的债券标准回购业务，参与者包括个人投资者和企业
    3. 场外交易：银行间债券市场，参与者包括中国人民银行、商业银行、证券公司、基金管理公司等金融机构
3. 主要类型
    1. 按回购期限分，交易所挂牌的债券回购可以分为1d/2d/3d/4d/7d/14d/28d/91d/182d
    2. 按逆回购是否有权处置回购协议标的的债券划分：
        1. 质押式回购
        2. 买断式回购（开放式回购）
4. 回购协议利率的影响因素
    1. 质押证券的质量:流动性、信用
    2. 回购期限的长短
    3. 货币市场及其他市场的利率利率
5. 定价
    1. $回购价格=本金\times{(1+回购时应付的利率\times\frac{回购协议期限}{360})}$
6. 回购协议的风险
    1. 到期时，证券的出让方（正回购方）无法按约定的价格赎回，证券的受让让（逆回购方）只能保留作为抵押物的证券
    2. 所质押的债券实际的市场价值低于质押式回购的本息，或者质押债券的流动性较差不能够通过卖出变现对回购违约进行补偿。

## 债券的收益率

### 当期收益率
1. 债券的年利息收入与当前的债券市场价格的比率：$I=\frac{C}{P}$
2. 当期收益率没有考虑债权投资所获得的资本利得或损失，只是债券某一时期所获得的现金收入相较于债券价格的比率

### 即期收益率/到期收益率（YTM）
1. 到期收益率是可以使投资购买债券获得的未来现金流的现值等于债券当前价格的贴现率
2. 到期收益率与债券市价的关系：【固定利率估值DCF公式】
3. 到期收益率隐含两个重要假设：
    1. 投资者持有至到期
    2. 利息再投资收益率不变

### 到期收益率的影响因素
1. 票面利率,同比
2. 债券市场价格，反向
3. 再投资收益率

### 当期收益率与到期收益率的关系
1. 债券市场价格越接近债券面值，期限越长，则其当期收益率越接近到期收益率，反之，偏离
2. 不论当期收益率和到期收益率近似程度如何，当期收益率的变动总是预示到期收益率的同向变动；

## 债券的久期和凸性

### 久期
1. 概念：债券本息所有现金流的加权平均到期时间，即投资者收回其全部本金和利息的平均时间；
2. 麦考利久期：【公式】
3. 对于零息债券，其久期等于到期期限
4. 对于附息债券，在债券到期之前的每一次付息都会缩短加权平均到期时间，因此付息时间的提前或者付息金额的增加都会使债券的久期缩短。
5. 修正久期：【公式】，收益率变动一个百分点对应价格变动百分比
6. 在给定的收益率变动下，债券价格的百分比变化与修正久期变化方向相反，修正久期越大，由给定收益率变化所引起的价格变化越大。
7. 【案例】

### 凸性
1. 当债券的持有收益率增加或减少同样的数量，债券价格会呈现不同幅度的变化特征【附图】
2. 价格-收益率的曲率就是债券的凸性
3. 凸性意味着债券的价格-收益曲线的斜率随着收益率而变化
4. 对于凸性为正的债券（不含期权的债券都是正的凸性），收益率升高，斜率是较小的负值，久期估算价格的下跌幅度大于实际价格下跌幅度；反之，反之；
5. 凸性对投资者是有利的，在其他特性相同时，投资者应该选择凸性更高的债券进行投资；

### 久期和凸性的应用
1. 所得免疫
    1. 保证投资者有充足的资金可以满足预期现金支付的需求
    2. 对于养老基金、社保基金、保险基金等机构投资者具有重要的意义：
    3. 三种策略：
        1. 现金配比策略：限制性强，弹性很小，可能会排斥许多缺乏良好现金流量特性的债券
        2. 究其配比策略：只要求负债的久期和债券投资组合的久期相同即可，因而有更多的债券可供选择。但为了满足负债的需求，债券管理者可能不得不在价格极低是抛出债券
        3. 水平配比策略：在短期内运用现金配比策略，在较长的时期内运用久期配比策略。即具有了现金配比策略中的流动性强的优点，又具有了久期配比策略中的弹性较大的优点；
2. 价格免疫
    1. 由那些保证特定数量资产的市场价值高于特定数量负债的市场价值的策略组成；
    2. 使用凸性作为衡量标准，实现资产凸性与负债凸性相匹配；
    3. 同等价值情况下，高凸性资产较低凸性负债，无论利率如何变化，资产价格表现都要好于负债的表现；【附图】

## 影响债券价值的主要因素

1. 安全性
    1. 信用评级
    2. 信用利差：企业债和金融债都要缴纳利息税，但金融债是国家信用，收益率要低于企业债
2. 流动性
    1. 品种、市场、发行人
    2. 流动性溢价
3. 税收利差
    1. 都是国家信用，金融债收益率要高于国债
4. 收益性
    1. 到期收益率水平：市场利率提高，债券价格下跌，反之亦然
5. 期限/剩余期限
    1. 多数情况下，较长期限的债券收益率要高于较短期限的债券
    2. 儿剩余期限长的债券（非浮息债），久期一般也越长（但是两者不是线性关系）

## 影响市场利率的主要因素

### 宏观经济基本面（经济好债券熊市，反之反之）
1. 经济增速-GDP
    1. 经济增长的三驾马车：投资、消费、净出口
    2. 先行指标：工业增加值、PMI
    3. 参考指标：工业用电、贷款增速
2. 通货膨胀指标-CPI/PPI
    1. CPI的构成：【二分法：一类二类项目】
    2. 食品的影响：猪肉的扰动

### 政策面
1. 货币政策-利率（价格）、准备金率（数量），之前是价格，当前数量操作
2. 汇率政策
3. 公开市场操作
4. 贷款的窗口指导


### 资金面
1. 货币供应：公开市场净投放、外汇占款、财政存款
2. 货币增速：M1、M2、社会融资总量
3. 货币市场资金状况：Repo、Shibor、商业银行超储水平
4. 股票市场：资金的转移或分配

### 市场供求
1. 供应：国债、金融债发行计划、信用债发行额度。。。
2. 需求：银行（存贷比）、保险（保费增速）、基金（新基金发行）。。。
3. 供求结构：银行倾向于短久期，保险倾向于长久期。。。

### 投资者风险偏好变化
1. 当市场风险偏好降低，投资者会迅速减少所持有的风险资产的风险暴露，使持有资产的价值减少，反之反之

### 信用利差
1. 对于给定的非政府部门的债券、给定的信用评级，信用利差随着期限增加而扩大
2. 信用利差随着经济周期的扩张而缩小，随着经济周期的收缩儿扩张
3. 信用利差的变化本质上是市场风险偏好的变化，受经济预期影响
4. 从实际数据看，信用利差的变化一般发生在经济周期发生转换钱。因此，信用利差可以作为预测经济周期活动的指标

### 总结
1. 不同的因素在不同的时期，其影响程度不同
2. 很难说某个因素会起全部或决定性作用
3. 市场反映的更多是预期，而不是当下
4. 超预期的变化很重要，其他因素在预期之内，某个因素发生了超预期的变化，经常会对市场产生巨大的冲击

# 债券投资的风险

## 市场风险

### 利率风险
1. 指利率变动引起债券价格波动的风险
2. 债券的价格与利率呈反向变动关系：【DCF估值公式】
3. 固定利率债券和浮动利率债券对市场利率的敏感度不同：
>浮动利率债券的利息在付息日会根据约定的条款进行设定，从而在市场利率上升的环境中具有较低的利率风险，而在市场利率下行的环境中具有较高的利率风险
4. 【走势图例】

### 通货膨胀风险
1. 所有种类的债券都会面临通货膨胀风险，因为利息和本金都是不能够随通货膨胀水平变化而变化
2. 通胀压力增加，企业投资扩大再生产的动力会增加，市场风险偏好上升，持有债券的意愿下降；同事中央银行为抑制通胀会提高利率水平，采取紧缩的货币政策，从而导致债券价格下跌，市场收益率上升；【走势图例】
3. 浮动利率债券因为其利息是浮动的，而在一定程度上降低了通胀风险，但其本金可能遭受的购买力下降而带来的损失无法避免；【走势图例】

## 信用风险
1. 指债券发行人未按照契约的规定支付债券的本息，给债券投资者带来损失的可能性。即使债券发行人未真正违约，债券持有人仍有可能因为信用风险而遭受损失
2. 列示【不同评级公司的评级符号体系】
3. 【走势图例】

## 流动性风险
1. 债券流动性是指债券投资者将手中的债券变现的能力
2. 通常用债券的买卖价差的大小反应债券的流动性大小；
3. 债券的流动性风险是指未到期债券的持有人无法以市值，而只能以明显低于市场价值变现的风险；
4. 【wind终端截图案例】

## 再投资风险
1. 指在市场利率下行的环境中，附息债券收回的利息或者提前于到期日收回的本金，只能以低于原债券到期收益率的水平再投资于相同属性的债券，而产生的的风险；

## 提前赎回风险
1. 指债券发行者在债券到期日前赎回有提前赎回条款的债券所带来的风险；
2. 债券发行人通常在市场利率下降时执行提前赎回条款，因此投资者只好将本息再投资于其他利率更低的债券，导致再投资风险；
3. 可赎回债券和大多数住房贷款抵押支持证券允许债券发行人在到期日前赎回债券，此类债券面临提前赎回风险；

# 衍生产品

## 衍生品基础
衍生品是由另一种基础资产构成或衍生而来的交易合约。

### 基础资产
股票、债券、货币等金融资产，也可以是黄金、原油等大宗商品。

### 衍生品五要素：
1. 基础资产/标的资产
2. 到期日
3. 交易单位
4. 交割价格
5. 结算

### 特点
1. 跨期
2. 杠杆
3. 联动
4. 可用于管理风险，自身也带有高风险

### 分类
1. 按合约：远期、期货、期权、互换、结构化金融工具
2. 按标的：货币、利率、权益、信用、商品、其他
3. 按场所：交易所交易-场内、场外

### 功能
1. 风险管理：套期保值
2. 价格发现：有套利保证、众多投资人众多信息的反应
3. 投资投机：消费风险

## 远期合约
1. 概念：交易双方约定在未来特定时间，按约定价格交易约定数量的标的资产
2. 特点：非标准化、契约可定制、市场效率低
3. 【案例：远期利率协议-FRA】

## 期货合约
与远期相似，买卖双方约定在未来特定时间按约定价格交易某项合约标的的资产

期货是标准化交易，在交易所场内进行

### 交易制度：
1. 保证金
2. 盯市制度
3. 对冲平仓
4. 交割制度

### 合约要素
1. 期货品种
2. 交易单位
3. 最小变动单位
4. 单日波幅限制
5. 合约到期月份
6. 交易时间
7. 最后交易日
8. 交割等级

【合约举例：沪深300股指期货】

## 期权
期权合约的买方有权利但没有义务在规定期限内按约定价格买入或者卖出约定数量的标的资产

多数在交易所场内交易

### 合约要素
1. 标的资产
3. 买方
2. 卖方
4. 执行价格
5. 期权费
6. 通知日
7. 到期日

### 分类
1. 美式期权和欧式期权
2. 看涨、看跌
3. 实值、平价、虚值

【上证50ETF期权】

交易双方在未来某一时间相互交换某种合约标的资产
非标准化，场外交易

### 种类
1. 利率
1. 货币
1. 股票收益
