# 报表分析：功力在表外

## 会计报表是会计师做出来的


### 从几个问题说开去
1. 货币政策、股市行情、是否会影响融资机会和会计报表？
2. 人民币持续升值，是否会影响轿车企业出口业务的毛利率？
3. 宝马等知名品牌大举进军中国市场，江淮等载货汽车企业进军乘用车市场，对轿车市场的竞争格局、对其他汽车企业的利润影响如何？
4. 国际原油价格超过135美元/桶，对汽车需求产生何种影响？夏利汽车与红旗汽车的毛利率孰高孰低？
5. 不同的折旧方法是否影响企业的产品成本和销售成本？

### 概言之
一定要结合企业所处的经济环境、行业环境、企业自身的竞争战略、企业选用的会计方法来理解影响企业业绩的诸因素及其未来发展。报表分析的功力在报表之外

财务报表数据的决定模型：
1. 宏观环境（7%影响权重）影响：产品结构（50%）、经营决策、，40%受经济背景
2. 行业环境（31%）影响:融资决策、会计方法（10%）

## 案例分析（歌华有线）
1. 03年净资产报酬率为什么会增加?
    1. 业绩变动：卖方垄断与定价
    2. 产业链条：唯一买方，唯一卖方
    3. 03年初，歌华有线先是关闭了开路信号，强化了产业链条下游的垄断地位
    4. 将有线电视收费从12元/月提高至18元/月
    5. 有线电视是必需品，需求量对价格不敏感，提价之后销量不变，收入增加
    6. 成本不变
    7. 股东权益不变，净资产报酬率增加
    8. 报表分析的功力在表外
2. 03年为什么货币资产增加，而不投资出去？
    1. 03年广电提出大力发展数字电视产业
    2. 为什么不投资？
    3. 行业背景，央视和歌华，形成了双头垄断，央视和歌华抢电视信号加密和用户收费权限
    4. 产业链条争夺，导致投资不执行
    5. 北京06年数字电视由于奥运会，歌华受政府压力才铺开

# 理解企业的现金流

## 现金流量表概述

### 现金的含义
1. 既包括库存的现钞，也包括可以开支票支付的银行存款。现金流入就是收到现金，现金流出就是付出现金。
2. 通常还包括少量的交易性金融资产

### 四类现金流量
1. 融资现金流
    1. 股权融资现金流量，优先级现金流（股票的发行现金流入，发放股利现金流出）
    2. 债券融资现金流量，劣后级现金流（债券发行，本息支付）
2. 经营活动现金流
    1. 成本: 薪酬、原材料采购等
    2. 收入：销售产品
3. 投资现金流，一般为负值（负债加杠杆），投资的目的是为了经营
    1. 资产置办：固定、无形资产置办付出现金
    2. 资产处置：折旧、摊销收到现金

### 两种编制方法
1. 直接法、间接法（上市公司现金流量一般是间接法，更好的揭示现金流）
2. 区别在于经营活动现金净流量的报告方式
2. 直接法下，直接报告销售商品收到的现金、采购支付的现金、为员工支付的现金等。即直接的报告现金流入流出和净流量。
3. 间接法，则是报告利润，扣除没有收到现金的的赊销收入，加上没有支付现金的赊购或费用，将净利润调整为现金净流量。换言之，间接法关注类似问题：
    1. 盈利企业一定有现金流？
    2. 亏损企业一定没有现金流？

## 间接法现金流量表的原理
1.  
    1. 经营活动现金流=销售收入-成本支付
    2. 经营活动现金净流量=净利润+折旧，亏损企业可能有正的现金流
2. 应收款的变动
    1. 经营活动现金净流量=净利润+折旧-应收款，盈利企业可能负的现金流，现金滞留在客户处
3. 存货的变动
    1. 经营活动现金净流量=净利润+折旧-存货增加
4. 应付账款变动
    1. 经营活动现金净流量=净利润+折旧+应付款

## 盈余管理：识别与案例
净利润=经营活动中现金净流量-折旧与摊销-财务费用-出售长期资产的损失+出售长期资产的收益+应收账款的增加+存货的增加-应付账款的增加-预收账款的增加

经营活动中的现金净流量=净利润+负债-资产

# 价值、清偿能力与投资回报分析

## 企业价值与净资产报酬率
1. 市净率=公司的市场价值/公司的净资产=每股市场价格/美股净资产
2. 净资产报酬率（ROE）=利润/净值产
3. 市净率和ROE成正比

## 投资报酬率与清偿能力
1. 投资回报率（=净利润/净资产）>资本成本，才具有清偿能力

## 分析净资产报酬率
1. 杜邦分解：ROE=净利润/净资产=净利润/销售收入X销售收入/总资产X总资产/净资产=销售净利润率X总资产周转率或周转次数X财务杠杆或权益乘数
2. 三部分比率都受企业经营环境（会计分析四要素）：
    1. 经济背景分析
    1. 行业背景分析
    1. 张略与运营分析
    1. 会计政策分析

## 案例分析（使用杜邦分解+四要素分析，对比分析）
1. 重庆路桥、其他公路上市公司：销售利润率、路和桥、替代品、议价能力
2. 贵州茅台、五粮液：高销售毛利润率
3. 联想集团、苹果：高销售利润率
4. 双汇发展、食品行业：行业性质决定议价能力不强。高资产周转次数、12年董事会报告：将信息化引入生猪屠宰、流程再造、订单采购、生产、销售，自办物流，自动化
