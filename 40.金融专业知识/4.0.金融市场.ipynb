# 市场组织和结构
## 金融系统的主要功能
1. 允许参与者在现在或将来借贷货币、募集资本、管理风险、交易资产，并且按照资产预估值来进行交易；
2. 确定使得市场资金供给和市场资金需求平衡的利率；
3. 将资本配置到最有效率的领域；

综上，金融市场使得资产和风险进行对象间和时间上的转移。其中交易方包括个人、公司、政府、慈善机构等。
## 金融系统功能的实现方式
1. 存款
2. 借债
3. 发行股票
4. 管理风险
5. 交换资产
6. 利用信息
7. 收益率的确定 
8. 资本配置

# 证券市场指数
证券市场指数，代表了一类资产、证券市场或子市场的表现，通常由一系列证券（叫做成分证券）组成。指数的数值由某一时点成分证券的市价计算得出，对应指数收益率是一段时间内指数的变化。
## 指数的构造和管理
1. 哪个目标市场的指数；
2. 包含目标市场中的哪些证券作为成分证券；
3. 如何确定权重；
4. 需要从新调整的频率；
5. 成分证券的选择和权重何时需要再检验；

## 指数构造不同的加权方法
1. 价格加权指数，成分证券价格的算数平均，例，道琼斯平均工业指数（DJIA）和日经道琼斯平均指数；
>$\large P=\frac{\sum_i^n{p_i}}{n}$
2. 等权重指数，成分证券收益率的算数平均，例，等金额持有每只成分证券，例如，标普500等权重指数;
>$\large P=P_0\times(\frac{\sum_i^n{r_i}}{n}+1)$
3. 市值加权指数，依据证券市值占比确定组合中的权重，例，上证综指；
>$\large P=\frac{\sum_i^n{p_i*v_i}}{P_0}$
4. 流通市值加权指数，例，沪深300；
>$\large P=\frac{\sum_i^n{p_{i}*v_{i\_flow}}}{P_0}$
5. 基本面加权指数，以及基本面指标占比确定组合中的权重，例，MSCI GDP 加权指数；
>$\large P=\frac{\sum_i^n{p_i*x_i}}{P_0}$

## 指数组合的调整和重构
1. 重新调整，当价格变动影响权重时，需要重新调整股票权重到目标权重。
2. 指数重构，指定期的将股票调进或调出指数。


## 证券指数的作用
1. 市场投资者情绪的反映
2. 基金经理的业绩基准
3. 衡量市场收益和风险
4. 衡量贝塔和风险调整后的收益率
5. 为指数基金的构造提供参照的投资组合

## 证券类指数的类型
1. 一般市场指数，例，上证综指、深证成指
2. 多市场指数，例，新兴市场指数、全球指数
3. 基本面加权的多市场指数，例，GDP加权全球指数
4. 板块指数，例，医疗、金融行业指数
5. 风格指数，例，价值、成长型指数

## 固定收益指数
1. 证券数量多
2. 交易商市场和交易不频繁

由于债券数量众多，使得指数中债券的数量有很大的差异。流动性不足、交易成本和成分股频繁变动导致基金经理复制债券指数比较困难、成本高。

# 市场有效性
信息有效市场，价格能够充分、快速、合理反应所有公开信息，即，目前价格是对所有信息的无偏估计，预期收益是对所承担风险的合理补偿，在这种情况下，打败市场是不可能的。

在完全有效市场中，投资者只能运用被动投资策略，即，买入市场指数并持有。

衡量市场有效性的方法，是价格反映信息所需要的时间，即滞后性。

## 市场价值和内在价值
市场价值就是资产的交易价格

内资价值或基础价值，是理性投资者基于所有信息和认识，愿意支付的价格；

如果市场高度有效，则投资者认为现价（交易价格）就是内在价值，如果市场不是完全有效，主动投资者可以买入他认为内在价值高于市价的资产，卖出认为内在价值地域市价的资产。

由于不同的投资者可能对同一资产有着不同的内在价值估计，内在价值并不确定，资产越复杂，估计内在价值越难，而且内在价值会随新的预期外信息而改变。

## 影响市场有效性的因素
1. 市场参与者数量
2. 信息的可获取性
3. 交易障碍
4. 交易成本和信息成本

## 有效市场假说
### 要点
1. 在市场上的每个人都是理性的经济人
2. 股票的价格反映了这些理性人的供求的平衡
3. 股票的价格也能充分反映该资产的所有可获得的信息

### 有效的资本市场的二种定义
1. 内部有效市场又称交易有效市场
2. 外部有效市场又称价格有效市场

### 成为有效市场的条件
1. 投资者都利用可获得的信息力图获得更高的报酬
2. 证券市场对新的市场信息的反应迅速 而准确，证券价格能完全反应全部信息
3. 市场竞争使证券价格从旧的均衡过渡到新的均衡，而与新信息相应的价格变动是相互独立的或随机的

### 有效市场假说的三种形态
1. 弱式有效市场
2. 半强式有效市场
3. 强式有效市场

### 市场有效性的检验
1. 弱式有效假说的检验
>序列自相关分析、技术分析是否能够获得超额收益
2. 半强式有效假说的检验
>事件检验法，即检验和公司基本面有关的事件发生时，股价变化有无快速反应。若能快速反应，则投资者不能通过新信息获得超额利润，基本分析失灵，半强式有效成立。
3. 强式有效假说的检验
>检验基金或有可能获得内幕消息人士的投资绩效评价，若被评估者的投资绩效确实优于市场平均，则强式有效不能成立。

## 对有效市场假说的挑战
### 面临的理论挑战
1. 投资者并非是完全理性的
2. 投资者不止偶然偏离理性，而是经常以同样的方式偏离理性
3. 套利者不会完全消除非理性投资者的错误对价格的影响

### 面临的实证挑战
1. 长期反转效应
2. 动量效应
3. 特殊时点
4. 小市值效应、pe/pb效应
5. 过度反应异常、反应不足异常、随机反应异常

## 行为金融学
1. 风险厌恶型投资者对于风险的喜好是不对称的
2. 投资者有时高估自己分析股票的能力
3. 最近的结果影响为对来的判断
4. 投资者将不同类型的资产分别对待而非视为整体
5. 对于变化的反应过慢
6. 愿意实现盈利，不愿意实现亏损
7. 将事件割裂开

### 分形市场假说的主要论点归纳

1. 当市场是由各种投资期限的投资者组成时，市场是稳定的。
2. 信息集对基本分析和技术分析来讲短期影响比长期影响要大。
3. 当某一事件的出现使得基础分析的有效性值得怀疑时，长期投资者或者停止入市操作或者基于短期信息进行买卖。
4. 价格是短期技术分析和长期基础分析的综合反应。
5. 如果某种证券与经济周期无关，那么它本身就不存在长期趋势。
