# 衍生品市场与工具
1. 衍生品，价值基于其他资产或证券收益的证券。
2. 交易所交易衍生品，是有清算中心担保的标准化的合约。
3. 场外交易（柜台交易），属于没有监管的市场，每一个合约都有一个对手方，持有人暴露在违约风险下。
4. 远期承诺，是一项有法律约束力的，约定未来采取某种行动的承诺，包括远期合约、期货合约、和互换。
5. 或有权利，是一项取决于特定事件的权利。期权是一项依赖于未来股价表现的或有权利。远期、期货和互换的兑付是基于一个确定的价格而无论价格的涨跌。
6. 远期合约，合同双方在未来某一时间以一定的价格，对一定量的实物资产或者证券进行交易约定。
7. 期货合约，在交易所交易的标准化的远期合约。
8. 互换合约，指一系列的远期合约组合。
9. 期权合约，指在未来一定时期可以买卖的权利，是买方向卖方支付一定数量金额（权利金）后，拥有的在未来一段时间内（美式期权）或未来某一特定日期（欧式期权），已实现约定好的价格（履约价格）向卖方购买（看涨期权）或出售（看跌期权）一定数量的特定标的物的权利，但不负有必须买进或卖出的义务。
10. 衍生品市场的作用：
>1. 价格发现
>2. 转移与分散风险；
>3. 降低交易成本；
11. 套利，本质就是无风险的超额收益，对应投资组合的无风险回报率高于市场无风险利率的投资机会。当资产的定价错误时，套利的机会出现，套利者会一直进行套利活动直到资产的价格恢复到有效水平。
>**两种套利理论：**
>1. 一价定律：两种资产或组合在将来有相同的现金流，则这两种资产或者组合的定价应该相同，否则可以套利。
>2. 两种资产将来的回报不确定，但组合有确定的回报率，组合的回报率应该等同于无风险利率，否则可以套利。

# 衍生品定价和估值
本节主要基于无套利定价，对衍生品定价和估值。

## 现金流
对于大多数风险资产来说，现值$S_0$等于资产在未来某时刻价格的折现值。因为未来价值是有风险的（收益不确定），折现率包括风险溢价和无风险收益率，即$R=R_s+R_f$。

$S_0=\frac{E(S_T)}{{(1+R_f+R_\alpha)}^T}\\S_0：资产即期价格\\E(S_T)：T时点预期价值$

以上，假设投资者是风险厌恶者，其对风险资产需要有一个正溢价补偿。如果一个风险中性投资者不要求风险溢价，结果是会将资产的期望值用无风险利率进行折现。

## 无套利
相对于现金流折现估值模型，对于衍生品证券的估值是基于无套利条件的。

无套利条件，是指由于市场定价的高效性，市场的套利机会很少，这是因为套利机会一旦出现，就会被迅速利用进行套利获利，进而消灭套利机会。

基于无套利条件，可以构建多组资产组合，无论发生什么，他们未来都有相同的回报。

对于衍生品，由于其风险完全基于相关资产的风险，所以，可以建立一个相关资产和衍生品的组合，这个组合的未来的价值就是确定的。因此，可以用无风险收益率折现计算组合的现值。无套利条件下的组合现值，使无风险组合的收益变为无风险收益。

${(S+F)}_0=\frac{{(S+F)}_T}{{(1+R_f)}^T}\\S:资产价值\\F:远期空头合约价值$

因为T（远期到期日）的收益是来自于一个完全对冲的组合，所以T时点的价值是确定的。无套利条件，是以上等式成立的基础。
>如果某时点$t$，购买资产的成本和卖出远期的的净成本小于$T$时点组合收益的现值（以$R_f$），那么投资者可以借资金（以$R_f$）买入这个组合，待市场进入无套利条件状态下（最晚持有至$T$时点），进行反向平仓（或到期交割），可以赚取超出$R_f$的无风险收益率。

>相反，如果净成本大于$T$时点组合收益的现值，那么投资者可以卖出组合（卖出资产，再以$R_f$进行无风险投资，然后买远期合约），待市场进入无套利条件状态下（最晚持有至$T$时点），进行反向平仓（或到期交割），可以赚取超出$R_f$的无风险收益率。


## 风险中性
在无套利条件下，投资者的风险厌恶并没有包含在衍生品的估值中。无套利衍生品定价也被称为风险中性定价。

## 复制
我们可以从一个基础资产和一个对冲的衍生品头寸来建立一个无风险资产（组合），那么：
1. 我们可以用无风险资产和基础资产复制出一个衍生品头寸的收益结构；
2. 或者可以用无风险资产和衍生品头寸复制一个基础资产的收益结构；

以上过程，称为复制，即，风险资产、无风险资产、衍生品头寸，其中任意两个可以组合成另外一个。

## 远期合约的价值和价格
因为买卖双方没有任何一方，在进入远期交易开始时会支付现金（实现亏损），所以远期合约的价格要使合约价值等于零（买卖双方都不会亏损）。资产在时点$0$的远期（$T$）价格为：

$F_0(T)=S_0{(1+R_f)}^T$

>如果远期价格是${\color{red}{F_0(T)+}}$，大于$S_0{(1+R_f)}^T$，套利者可以做空一个远期合约，承诺在$T$时点以${\color{red}{F_0(T)+}}$卖出资产，然后以$R_f$借款在$S_0$价格上买入资产。在时点$T$，套利者可以交付资产收到${\color{red}{F_0(T)+}}$，以$S_0{(1+R_f)}^T$作为成本偿付借款，那么获取${\color{red}{F_0(T)+}}$与$S_0{(1+R_f)}^T$正差值的收益；

>如果远期价格是${\color{green}{F_0(T)-}}$，小于$S_0{(1+R_f)}^T$，套利者以$S_0$价格卖出资产，将收入以$R_f$进行无风险投资，然后持有一个远期合约的多头。在时点$T$，套利者可以获得无风险投资收入$S_0{(1+R_f)}^T$，支出${\color{green}{F_0(T)-}}$购买资产来覆盖空头资产头寸，那么获得${\color{green}{F_0(T)-}}$与$S_0{(1+R_f)}^T$正差值的收益；

以上机制，保证远期合约的价格$F_0(T)$在无套利条件下使远期合约的价值在时点$0$是零价值，即，当远期报价是其无套利价格时，期初远期合约价值：

$V_0(T)=S_0-\frac{F_0(T)}{{(1+R_f)}^T}=0$

在期初至到期日$T$期间，时点$t<T$时，远期合约的价值：

$V_t(T)=S_t-\frac{F_0(T)}{{(1+R_f)}^{T-t}}$

特别的，在到期日$T$，折现式${(1+R_f)}^{T-t}=1$，所以，远期多头的收益是$S_T-F_0(T)$，即，资产到期日的即期价格和远期合约价格之差。

## 考虑持有资产的成本和收益
将时点$t-T$之间持有资产的所有成本的现值表示为$P_{c,t}$，期间资产产生的现金流和所有便利收益的现值表示为$P_{r,t}$。那么，无套利条件下：

$F_0(T)=(S_0+P_{c,0}-P_{r,0}){(1+R_f)}^T$

在$0-T$期间时点$t$，远期合约的价值：

$V_t(T)=S_t+P_{c,t}-P_{r,t}-\frac{F_0(T)}{{(1+R_f)}^{T-t}}$

特别的，在到期日$T$，持有资产到期的成本和收益都为零，远期多头的收益是$S_T-F_0(T)$。

## 期权
期权的价值状态，指的是期权是否处于价内或者价外。
>* 如果马上行权能产生正收益，是价内
- 如果马上行权产生负收益，是价外
- 如果不能产生正收益或者负收益，是平价

我们定义：$S$是基础资产的价格，$X$是期权的行权价格。那么：

**认购期权：**
>- 价内： $S>X$
- 价外：$S<X$
- 平价：$S=X$

**认沽期权：**
>- 价内： $S<X$
- 价外：$S>X$
- 平价：$S=X$
## 期权价值
$认购期权价值=\underbrace{max(S-X,0)}_{内在价值}+时间价值\\认沽期权价值=\underbrace{max(X-S,0)}_{内在价值}+时间价值$

期权时间价值：是指期权价超出内在价值的部分，也就做期权的投机价值。

## 决定期权价值的因素
1. 基础资产的价格，对认购正向，对认沽负向
2. 行权价，对认购负向，对认沽正向
3. 无风险收益率，对认购正向，对认沽负向
4. 基础资产的波动率，对认购和认沽都是正向
5. 距离到期日，对认购和认沽都是正向
6. 持有资产的成本和好处

## 欧式期权的看跌-看涨期权的平价关系
构建两个组合一个信托买入期权和一个保护看跌期权。

信托买入期权，是一个行权价是$X$的看涨期权和一个纯折现、无风险的到期日（也是期权到期日）支付X的债券组合。


保护看跌期权，是一个股票和基于其的看跌期权的组合。

如果到期时，$S\geq X$：
>- 保护看跌期权支付股票$S$，同时看跌期权到期是价值为零，所以收益为$S$。
- 信托买入期权支付债券$X$，同时看涨期权支付$S-X$，所以收益为$X+(S-X)=S$。

如果到期是，$S<X$：
>- 保护看跌期权支付股票$S$，同时看跌期权支付$X-S$，所以收益为$S+(X-S)=X$。
- 信托买入期权支付债券$X$，同时看涨期权到期是价值为零，所以收益为$X$。

如上，无论哪种情况，保护看跌期权和信托买入期权的收益是一样的。那么，基于无套利条件，可以把看跌-看涨期权平价关系表示为：

$\underbrace{c}_{看涨期权价值}+\underbrace{\frac{X}{{(1+R_f)}^T}}_{债券现值}=\underbrace{S}_{股票价格}+\underbrace{p}_{看跌期权价值}$
